## Merging Dataframes

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'/Users/biancaniemann/Documents/Alex_The_Analyst/Pandas_Section/LOTR.csv')
df2 = pd.read_csv(r'/Users/biancaniemann/Documents/Alex_The_Analyst/Pandas_Section/LOTR 2.csv')
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
# automatically an inner join and joins on both Fellowship ID and First name
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
#Here we can specifY what type of join and where to join
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
# basically the same as first merge just written out
```

```python
df1.merge(df2, how = 'outer')
# joins both together but puts NAN for missing info
```

```python
df1.merge(df2, how = 'left')
# pulls everything from df1 and only brings from df2 if it matches - basically just age here comes through and only 2 values
```

```python
df1.merge(df2, how = 'right')
#pulls everything from the right side here
```

```python
df1.merge(df2, how = 'cross')
#takes everything in right(df1) and compares to left(df2) for every single value
```

# joins

```python
df1.join(df2, on = ['FellowshipID'], how= 'outer', lsuffix = '_left', rsuffix = 'right')
#have to specify where to merge, how and the left and right suffixes
#join works better on indexes
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = 'right')
```

```python
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = 'right', how = 'outer')
```

```python
df4
```

# Concatanate

```python
pd.concat([df1, df2])
# puts one database ontop of the other and NAN for missing values
```

```python
pd.concat([df1, df2], join = 'inner')
# only shows where cols are the same
```

```python
pd.concat([df1, df2], join = 'outer')
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```
